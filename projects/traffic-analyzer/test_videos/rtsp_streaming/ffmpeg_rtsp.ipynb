# Запуск RTSP через FFMPEG

*FFMPEG* - набор свободных библиотек с открытым исходным кодом, которые позволяют записывать, конвертировать и передавать цифровые аудио- и видеозаписи в различных форматах. 

*RTSP (Real Time Streaming Protocol)* - это сетевой протокол, предназначенный для управления потоковой передачей данных, таких как аудио и видео.

### Запускаем RTSP стрим через ffmpeg:
- `-i $VIDEO_PATH` - путь до файла видеофайла (mp4/avi/mov)
- `$RTSP` - адрес стрима
- `-stream_loop -1` - зациклить видео n раз (-1 означает - бесконечный цикл)
- `-re` - будем использовать нативный FPS видео (количество кадров в секунду)
- `-f rtsp` - будем стримить через протокол RTSP
- `-rtsp_transport tcp` - в качестве транспортного протокола выбираем TCP (альтернатива - UDP)

```python
VIDEO_PATH="../test_video.mp4"    
RTSP="rtsp://127.0.0.1:8554/input"

! ffmpeg -re -stream_loop -1 -i $VIDEO_PATH  -f rtsp -rtsp_transport tcp $RTSP
```
